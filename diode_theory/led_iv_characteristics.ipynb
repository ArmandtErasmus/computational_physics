# Non-Linearity Between the Voltage and Current Measurements in Diodes

```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

Import the I-V measurements for different coloured LEDs.

```python
# import the data
r_path = r'data\red_led.csv'
o_path = r'data\orange_led.csv'
y_path = r'data\yellow_led.csv'
g_path = r'data\green_led.csv'
b_path = r'data\blue_led.csv'
red, orange, yellow, green, blue = pd.read_csv(r_path), pd.read_csv(o_path), pd.read_csv(y_path), pd.read_csv(g_path), pd.read_csv(b_path)
```

```python
# create a color mapping to match plots to color
color_mapping = {'red': red, 'orange': orange, 'yellow': yellow, 'green': green, 'blue': blue}

plt.figure(figsize=(12,6))
# loop through the plots for efficiency
for color_name, color_df in color_mapping.items():
    plt.plot(color_df['Measured Voltage [V]'], color_df['Measured Current [A]'], ls='-', lw=2, color=f'{color_name}', label=f'{color_name} LED')
    
plt.legend()
plt.xlabel('Forward Voltage')
plt.ylabel('Forward Current')
plt.title('Relationship Between Forward Voltage and Forward Current')
plt.show()
```
